## **1.**

**1)**

$ \| x \|_\infty = \max\limits_i |x_i|$

From one hand: 

$\| x \|_1 = \Sigma_{i = 1}^n|x_i| \leq n \cdot \max\limits_i |x_i| = n \cdot \| x \|_\infty$

$\Rightarrow \| x \|_1 \leq n \cdot \| x \|_\infty \Rightarrow \frac{1}{n} \cdot \| x \|_1 \leq \| x \|_\infty$

$\Rightarrow C_1 = \frac{1}{n}$

From the other hand:

$\| x \|_1 = \Sigma_{i = 1}^n|x_i| \geq \max\limits_i |x_i| = \|x\|_\infty$

$\Rightarrow \| x \|_1 \geq \| x \|_\infty$

$\Rightarrow C_2 = 1$

### Answer:
$C_1 = \frac{1}{n}, C_2 = 1$

### **2)**

Given matrices: $A$ - arbitrary, $U$ - unitary

We will use this property of unitary later: $U^*U = UU^*  = I$

$\|UA\|^2_F = tr[(UA)^* (UA)] = tr[A^*\underbrace{U^*U}\limits_{I}A] = tr[A^*A] = \|A\|_F^2$

Now we'll need trace property: $tr[A \cdot B] = tr[B \cdot A]$

Let's derivate it:

$tr[AB] = tr[ \{\sum_k a_{ik}b_{kj}\}_{i, j}] = tr[ \{\sum_k a_{ik}b_{ki}\}_i] = tr[ \{\sum_k b_{ki}a_{ik}\}_i] = tr[BA]$

$\Rightarrow tr[AB] = tr[BA]$

$\|AU\|^2_F = tr[(AU)^* (AU)] = tr[(AU)(AU)^*] = tr[A\underbrace{UU^*}\limits_{I}A^*] = tr[AA^*] = tr[A^*A] = \|A\|_F^2$
### Answer:
Thus, 

$\|A\|_F = \|UA\|_F = \|AU\|_F, \forall$ unitary $U$

### **3)**

Given matrices: $A$ - arbitrary, $U$ - unitary

$\|UA\|^2_2 = \lambda_{max}[(UA)^* (UA)] = \lambda_{max}[A^*\underbrace{U^*U}\limits_{I}A] = \lambda_{max}[A^*A] = \|A\|^2_2$

Let's derivate some rule thank will soon come in handy:

$AB(x) = \lambda (x)$ 

$\Rightarrow BA(Bx) = B\lambda (x) = \lambda (Bx)$

$\Rightarrow$ eigenvalues of $AB$ and $BA$ are the same.


$\|AU\|^2_2 = \lambda_{max}[(AU)^* (AU)] = \lambda_{max}[(AU) (AU)^*] = \lambda_{max}[A\underbrace{U U^*}\limits_{I}A^*] = \lambda_{max}[AA^*] = \lambda_{max}[A^*A] = \|A\|^2_2$
### Answer:
Thus, 

$\|A\|_2 = \|UA\|_2 = \|AU\|_2, \forall$ unitary $U$

### **1)**

Using SVD: $A = U\Sigma V^*$,

where:

$\Sigma$ is diagonal matrix with non-zero elements $\sigma_{max} = \sigma_1 >= \sigma_2 >= ... >= \sigma_k$ on diagonal, $U$ and $V$ are unitary matrixes. $Rank(A)$ is amount of non-zero sigular values of $A$

$\|A\|^2_F = tr[A^*A] = tr[(U\Sigma V^*)^*(U\Sigma V^*))] = tr[V\Sigma^*\underbrace{U^* U}\limits_{I}\Sigma V^*)] = tr[V\Sigma^*\Sigma V^*)] = tr[(\Sigma V^*)^*(\Sigma V^*)] = tr[(\Sigma V^*)(\Sigma V^*)^*] = tr[\Sigma \underbrace{V^*V}\limits_{I}\Sigma^*] = tr[\Sigma\Sigma^*] = tr[\Sigma^*\Sigma] \leq rank(A) \cdot \sigma_{max}(A) = rank(A) \cdot \sigma_1(A) = rank(A) \cdot \|A\|_2^2$


### Answer:

Thus,

$\|A\|_F \leq \sqrt{rank(A)} \cdot \|A\|_2$



### **2)**

Let's prove that $\forall m, n, k: k \leq min(m, n), k = rank(A) \rightarrow \exists A \in \mathbb{R}^{m \times n}:$
$\|A\|_F = \sqrt{rank(A)} \cdot \|A\|_2$ 

$\triangleright$

The case, when $n = m$ is simplified general in some sense, because $k$ is less or equal to the smallest dimension in A. Thus, there is no need to describe extra cases with rectangular matrices.

Let's consider $A = I_n$ with 1 zero on it's diagonal:

$A = 
\begin{pmatrix}
  0 & 0 & \cdots & 0 \\
  0 & 1 & \cdots & 0 \\
  \vdots  & \vdots  & \ddots & \vdots  \\
  0 & 0 & \cdots & 1
 \end{pmatrix}$

$\|A\|_F = \sqrt{n - 1}$

$rank(A) =  n - 1$

$\|A\|_2 = \sigma_1(A) = 1$

$\Rightarrow \|A\|_F = \sqrt{n - 1} = \sqrt{rank(A)} \cdot \sigma_1(A) = \sqrt{n - 1} \cdot 1 = \sqrt{n - 1}$

### Answer:

Thus, $\forall n$ we could make such matrix $A$, that would consist of identity $I_{n-1}$ and would satisfy the equality above. $\triangleleft$ 

### **3)**

Let's apply exponent first:

$exp(A) = P \cdot exp(D) \cdot P^{-1}$

Now let's apply determinant with property $det(AB) = det(A)det(B)$ and $det(A^{-1})\cdot det(A) = 1$:

$det[exp(A)] = det[P \cdot exp(D) \cdot P^{-1}] = \underbrace{det(P)\cdot det(P^{-1})}\limits_{1}\cdot det[exp(D)] = \prod\limits_{i = 1}^n e^{d_i} = exp[\sum\limits_{i=1}^nd_i] = exp[tr(D)] = exp[tr(P^{-1}AP)] = exp[tr(A\underbrace{PP^{-1}}\limits_{I})] = exp[tr(A)]$

### Answer:

Thus, we proved, that:

$det[exp(A)] = exp[tr(A)]$

### **4)**

We are to prove, that: $\| AB \|_2 \leq \| A \|_2 \| B \|_F$

$\triangleright$

$\| AB \|_2 = \sup\limits_{\| x \|_2 \neq 0} \frac{\| ABx \|_2}{\| x \|_2} = \frac{\| ABx \|_2}{\| Bx \|_2}\cdot \frac{\| Bx \|_2}{\| x \|_2} \leq \| A \|_2 \cdot \| B \|_2$

From the other hand,

Let's estimate the lower bound of Frobenius norm $\|B\|_F = \sqrt{tr(B^*B)} \geq \sigma_1(B)$

Hence, $\|B\|_2 \leq \|B\|_F$

And it gives us:

$\| AB \|_2 \leq \| A \|_2 \cdot \| B \|_2 \leq \| A \|_2 \cdot \| B\|_F \triangleleft$

Here we will use following facts:

Let's open the brackets:

$F(U, V)) = \frac{1}{2}tr[(X - UDV)^T(X - UDV)] = \frac{1}{2}[tr(X^TX) - tr(X^TUDV) - tr(V^TDU^TX) + tr(V^TDU^TUDV)]$

$\frac{\partial F}{\partial V}(U, V) = \frac{1}{2}[\underbrace{\frac{\partial F}{\partial V}tr(X^TX)}\limits_{0} - \frac{\partial F}{\partial V}tr(X^TUDV) - \frac{\partial F}{\partial V}tr(V^TDU^TX) + \frac{\partial F}{\partial V}tr(V^TDU^TUDV)] = \frac{1}{2}[- \frac{\partial F}{\partial V}tr(X^TUDV) - \frac{\partial F}{\partial V}tr((X^TUDV)^T) + \frac{\partial F}{\partial V}tr(V^TDU^TUDV)] = \frac{1}{2}[- 2\frac{\partial F}{\partial V}tr(X^TUDV)+ \frac{\partial F}{\partial V}tr(V^TDU^TUDV)] = \frac{1}{2}[- 2DU^TX + \frac{\partial F}{\partial V}tr(V^TDU^TUDV)]$

$\frac{\partial F}{\partial V}tr(V^TDU^TUDV) = \frac{\partial F}{\partial V}tr(\overset{\downarrow}{V^T}DU^TUDV) + \frac{\partial F}{\partial V}tr(V^TDU^TUD\overset{\downarrow}{V})$, arrow denotes V, that we treat like a variable

$\Rightarrow \frac{\partial F}{\partial V}tr(V^TDU^TUDV) = \frac{\partial F}{\partial V}tr(V^TDU^TUD\overset{\downarrow}{V}) + \frac{\partial F}{\partial V}tr(V^TDU^TUD\overset{\downarrow}{V}) = 2\frac{\partial F}{\partial V}tr(V^TDU^TUD\overset{\downarrow}{V}) = 2DU^TUDV$

$\Rightarrow \frac{\partial F}{\partial V}(U, V) = -DU^TX + DU^TUDV = DU^T(-X + UDV)$

**2)**

````python
import jax
import jax.numpy as jnp

@jax.jit
def f(X, U, D, V):
  res = jnp.linalg.norm(X - U @ D @ V, ord='fro') 
  return 1/2 * res ** 2

@jax.jit
def agrad_u(X, U, D, V):
  return (-X + U @ D @ V) @ V.T @ D

@jax.jit
def agrad_v(X, U, D, V):
  return D @ U.T @ (-X + U @ D @ V)
````

````python
grad_f_u = jax.grad(f, argnums=1, has_aux=False)
grad_f_v = jax.grad(f, argnums=3, has_aux=False)

jit_grad_u = jax.jit(grad_f_u)
jit_grad_v = jax.jit(grad_f_v) 
````

### Second step: neural network architecture

For simplicity and demonstration purposes of the neural network compression idea consider the architecture consisting of the only fully-connected layers and non-linear ReLU functions between them.
To demonstrate compression effect, consider the dimension of the inner layers equals to 1000.

Below you see implementation of such neural network in PyTorch.
More details about neural networks you will study in the *Deep learning* course in one of the upcoming term

````python
class ResidualBlock(nn.Module):
    def __init__(self, in_features, middle_features, activation=nn.ReLU()):
        super().__init__()
        self.layer1 = nn.Linear(in_features, middle_features)
        self.layer2 = nn.Linear(middle_features, in_features)
        self.activation = activation
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.activation(out)
        out = self.layer2(out)
        out = out + x
        out = self.activation(out)
        return out

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.res1 = ResidualBlock(1000,1000)
        self.res2 = ResidualBlock(1000,1000)
        self.fc2 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.res1(x)
        x = self.res2(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.train()
    model.to(device)
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    device = "cuda" if torch.cuda.is_available() else "cpu"
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
    
    return 100. * correct / len(test_loader.dataset)
````

### Set parameters for training and print intermediate loss values

### Third step: run training with the [Adam](https://arxiv.org/pdf/1412.6980.pdf%20%22%20entire%20document) optimization method

````python
model = Net()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(1, epochs + 1):
    train(model,  train_loader, optimizer, epoch)
    test(model, test_loader)
````

````python
class MyResidualBlock(nn.Module):
    def __init__(self, res_block, r):
        super().__init__()
        self.layer1_UD, self.layer1_V = self.get_compressed_layer(res_block.layer1, r)
        self.layer2_UD, self.layer2_V = self.get_compressed_layer(res_block.layer2, r)
        self.activation = res_block.activation
        
    def get_compressed_layer(self, layer, r):
        new_layer_V = nn.Linear(r, 1000, bias=False)
        new_layer_UD = nn.Linear(1000, r, bias=False)

        U, D, V = torch.svd(layer.weight)

        new_layer_V.weight = nn.Parameter((V.T[:r]))
        new_layer_UD.weight = nn.Parameter(((U[:, :r]) @ torch.diag(D[:r])))
        new_layer_UD.bias = nn.Parameter(layer.bias) 

        return new_layer_UD, new_layer_V
    
    def forward(self, x):
        out = self.layer1_UD(self.layer1_V(x))
        out = self.activation(out)
        out = self.layer2_UD(self.layer2_V(out))
        out = out + x
        out = self.activation(out)
        return out

class MyNet(nn.Module):
    def __init__(self, model, r):
        super(MyNet, self).__init__()
        self.fc1 = model.fc1
        self.res1 = MyResidualBlock(model.res1, r)
        self.res2 = MyResidualBlock(model.res2, r)
        self.fc2 = model.fc2
        self.ReLU = model.ReLU

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.res1(x)
        x = self.res2(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
````
