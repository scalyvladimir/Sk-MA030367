# Black-Scholes Model

## 1. Practical relevance

The list of the main areas of use includes:

- search for undervalued options for sale
- identification of overvalued options for acquisition
- portfolio hedging in order to reduce risks in conditions of low volatility
- forecasting the market by the future value of the indicator of variability

The model is actively used by traders. It allows them to perform a comparison of theoretical and current price option values. If the theoretical and current parameters do not match, the arbitrage tactics are activated on this difference. Despite the fact that the pricing formula does not involve arbitration practice, in fact, several players use the model, finding and displacing arbitration situations in the market.

## 2. Thereotical properties of the system

The **B-S** model describes the behavior of options. But what is an *option* by itself?


***Option*** - a contract which gives an *ability*, **NOT** a *commitment* to buy/sell the asset with the price, which was agreed in advance and and called ***strike***.

Furthermore, for simplicity, let's consider a **call-option**, *payout function* of whose could be expressed with the following formula:

$$C = max(x - x_s; 0)$$
Where:

$x$ - base price of an asset,

$x_s$ - strike price.

The fairness of this contract is achieved in case that before making the deal, both sides are in same conditions and nobody knows, how the option will behave in the future. Considering this assumption, we will think of $C$ like it is a stochastic process depending on price of underlying asset $x$ and time $t$.

Now, when we assume that **B-S** formula will somehow depend on some stochastic variable, let's derivate some auxiliary formulas.

### Derivation of the Ito's lemma

A very first approximation of how price changes through time is: $$d x=r x d t$$
    
Where:

$x$ - asset price,

$t$ - time,

$r$ - risk-free rate, considering inflation and etc. .

The solution looks pretty simple: 
    $$x=x_{0} e^{r t}$$
    
Where:

$x_{0}$ - asset price at $t = 0$.

#### Random nature

Think of economy as a complex system, where any factor represented as a elemetary particle which could influence some chosen yellow particle, our option price.

<img src="pics/brown.gif" width="300" height="0" align="center"/>

At each moment of time $t_n$, a very large number of molecules have a multidirectional effect on the particle, while the force of their collision with the particle is also different. As a result, the observed particle makes chaotic movements. Such a picture is typical for the financial market.

<img src="pics/rand_vars.png" width="300" height="240" align="center"/>

If we transfer Brownian motion to a coordinate plane and represent it in discrete time, we get a Wiener variable describing one specific implementation of a random process: 
$$W_t = \varepsilon_1 + \varepsilon_2 + \cdots+\varepsilon_n$$

Where $\varepsilon_j$ are independent random variables having a normalized normal distribution $\sim N(0,1)$.

If $\varepsilon_j$ is an independent variable, we could descibe it with some statistics, such as correlation. In finance theory it is also called ***volatility***, introduced with $\sigma$ coefficient.

As a result, the discrete Wiener process is transformed into the formula:

$$W_t =\sigma( \varepsilon_1 + \varepsilon_2 + \cdot+ \varepsilon_n)$$

Due to the properties of random variables distributed normally, the sum of the Gaussian numbers $\varepsilon_1 + \varepsilon_2 +\cdots+\varepsilon_n$ is represented as $\varepsilon{\sqrt{n}}$, where $\varepsilon_j \sim N(0,1)$, and n is the total number of random price movements.

Ultimately, the Wiener process can be represented as $W_t =\sigma\varepsilon {\sqrt{n}}$. We also could interpret ${\sqrt{n}}$ as $\sqrt{\Delta t}$, that we will use later.

### Thus, we got our second approach:

$$x=x_{0}+r \Delta t+\sigma \varepsilon \sqrt{\Delta t}$$

### But reallity is little bit more different

Back in the 30s, it was found that it was not the prices themselves that were normally distributed, but their logarithms. Therefore, the object of study should not be himself, but $dx$. Below we will consider the logic of the transition from the cheka to $d(ln x)$.

Let's refine our model

At first, make a transition to differentials:
    $$x=x_{0}+r \Delta t+\sigma \varepsilon \sqrt{\Delta t} \mapsto \Delta x=r \Delta t+\sigma \varepsilon \sqrt{\Delta t} \mapsto d x=r d t+\sigma \delta W$$

Equation $d x=r d t+\sigma \delta W$ by itself an *Ito-process*, whose general view: $$d x=a(x, t) d t+b(x, t) \delta W$$

Where:

$a(x, t)$ - "demolition" function,

$b(x, t)$ - volatility function.

Bearing in mind, that we wanted to analyze $d(lnx)$ we also need one more transition:
$$d x=x r d t+x \sigma \delta W$$

And it's a lot more complex equation and in order to find its solution we should consider functional depending on stochastic function $x(t)$, which reminds us about our goal in the very beginning.

Thus, we have a general Ito process:

$$d F(x(t), t)=A(x, t) d t+B(x, t) \delta W$$

Let's use Ito to find $A(x, t) d t$ and $B(x, t)$ and then we'll discuss which condition F functional should satisfy.

Applying Taylor's expansion:

$$F(x, t)=F\left(x_{0}, t_{0}\right)+\frac{\partial F}{\partial x} \Delta x+\frac{1}{2} \frac{\partial^{2} F}{\partial x^{2}}(\Delta x)^{2}+\frac{\partial F}{\partial t} \Delta t+\cdots$$

Where:

Partial derivatives calculated at the point $(x_{0}, t_{0})$

$\Delta x=x-x_{0}, \Delta t=t-t_{0}$

We also will consider only infinitesimal of $\Delta x, \Delta t,(\Delta x)^{2}$ the order of smallness.

Calculating Expectancy:

$$\langle\Delta F\rangle=\frac{\partial F}{\partial x}\left(a_{0} \Delta t\right)+\frac{1}{2} \frac{\partial^{2} F}{\partial x^{2}}\left(b_{0}^{2} \Delta t\right)+\frac{\partial F}{\partial t} \Delta t+\cdots$$

In case that:
   
$\langle\varepsilon\rangle=0$,  
$\langle\Delta x\rangle=a_{0} \Delta t+b_{0}\langle\varepsilon\rangle \sqrt{\Delta t}=a_{0} \Delta t$, 
    
$\langle\varepsilon^{2}\rangle=1$,

$(\Delta x)^{2}=a^{2} \Delta t^{2}+2 a b \varepsilon \Delta t \sqrt{\Delta t}+b^{2} \varepsilon^{2} \Delta t =b_{0} \Delta t$

And so, we got:

$\begin{equation}
    \begin{cases}
    A(x, t)=\frac{\langle\Delta F\rangle}{\Delta t}\\
    B(x, t)=\frac{\left\langle\Delta F^{2}\right\rangle}{\Delta t}
    \end{cases}\,
\end{equation}$

After substituting F:

$\begin{equation}
    \begin{cases}
    A(x, t)=\frac{\partial F}{\partial x} a_{0}+\frac{b_{0}^{2}}{2} \frac{\partial^{2} F}{\partial x^{2}}+\frac{\partial F}{\partial t}\\
    B(x, t)=b_{0}\left(\frac{\partial F}{\partial x}\right)
    \end{cases}\,
\end{equation}$

$d F(x(t), t)=A(x, t) d t+B(x, t) \delta W = \left(\frac{\partial F}{\partial t}+a(x, t) \frac{\partial F}{\partial x}+\frac{b^{2}(x, t)}{2} \frac{\partial^{2} F}{\partial x^{2}}\right) d t+b(x, t) \frac{\partial F}{\partial x} \delta W$

Thus, $F(t, x) \in C^{1,2}\subset \mathbb{R}_{+} \times \mathbb{R} $

#### Logarithmic walk

$F = lnx$

$d(\ln x)=\left(\frac{\partial(\ln x)}{\partial t}+r x \frac{\partial(\ln x)}{\partial x}+\frac{(\sigma x)^{2}}{2} \frac{\partial^{2}(\ln x)}{\partial x^{2}}\right) d t+\sigma x \frac{\partial(\ln x)}{\partial x} \delta W = \left(0+r x \frac{1}{x}-\frac{(\sigma x)^{2}}{2} \frac{1}{x^{2}}\right) d t+\sigma x \frac{1}{x} \delta W$

$\Rightarrow d(\ln x)=\left(r-\frac{\sigma^{2}}{2}\right) d t+\sigma \delta W$

#### Finaly got derivated equation of price evolution:

$$x=x_{0} e^{\left(r-\frac{\sigma^{2}}{2}\right) \Delta t+\sigma \delta W}$$

As a result, we come to a fundamentally important equation that underlies most models for assessing the fair value of financial instruments.

```python
import numpy as np
import warnings
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning) 
import matplotlib.pyplot as plt
import scipy.stats


plt.style.use('ggplot')
plt.rcParams['lines.linewidth'] = 0.6

fig, ax = plt.subplots()
fig.set_figwidth(22)
fig.set_figheight(15)

colormap = plt.cm.gist_ncar

r = 0.25
sigma = 0.1
dt = 1 / 360
x0 = 100
num_plots = 5
num_days = 360 * 8

history = [x0]
days = [0]

for i in range(1, num_days + 1):
    days.append(i)

for _ in range(num_plots):
    for _ in range(num_days):
        
            eps = scipy.stats.norm(loc=0, scale=1).rvs(size=1)
            
            val = history[-1] * np.exp((r - sigma**2 / 2) * dt + (sigma * eps * np.sqrt(dt)))
            
            history.append(val)

    plt.plot(days, history)
    history = [x0]
plt.title('Evolution of price in time')
plt.savefig('price_dist.png')
pass
```

### Back to the Option

Recalling what we received earlier:

$\begin{equation}
    \begin{cases}
    C = max(x - x_s; 0)\\
    d x=x r d t+x \sigma \delta W \\
    d F(x(t), t)= \left(\frac{\partial F}{\partial t}+a(x, t) \frac{\partial F}{\partial x}+\frac{b^{2}(x, t)}{2} \frac{\partial^{2} F}{\partial x^{2}}\right) d t+b(x, t) \frac{\partial F}{\partial x} \delta W
    \end{cases}\,
\end{equation}$

Substituting $C$ in $dF$:

$d C=\left(\frac{\partial C}{\partial t}+x r \frac{\partial C}{\partial x}+\frac{\sigma^{2}}{2} \frac{\partial^{2} C}{\partial x^{2}}\right) d t+x \sigma \frac{\partial C}{\partial x} \delta W$

To solve this complicated equation we'll need t use transformation $\Pi=\frac{\partial C}{\partial x} \cdot x-C(x, t)$.

$$\frac{\partial C}{\partial \tau}+r C=\frac{\sigma^{2} x^{2}}{2} \frac{\partial^{2} C}{\partial x^{2}}+r x \frac{\partial C}{\partial x}$$

Where $\tau = T - t$, $T$ denotes time period.

Last representation can be transformed into the equation of thermal conductivity, solution of those will give our goal law:

$$C=x_{0} \mathcal{N}\left[d_1\right]-x_{s} e^{-r \tau} \mathcal{N}\left[d_2\right]$$

Where:

$\mathcal{N}$ - normal distribution function,

$\sigma$ - periodic volatility,

$d_1 = \frac{\ln \left(x e^{r \tau} / x_{s}\right)}{\sigma \sqrt{\tau}}+\frac{\sigma \sqrt{\tau}}{2}$,

$d_2 = \frac{\ln \left(x e^{r \tau} / x_{s}\right)}{\sigma \sqrt{\tau}}-\frac{\sigma \sqrt{\tau}}{2}$.

## 3. Modeling

Derivated law has a number of assumptions:

1. Interest rate is known and constant through time. 

2. The stock follows a random walk in continuous time, the variance of the stock price paths follow a log-normal distribution. 

3. Volatility is constant 

4. Stock pays no dividends (can be modified to include them however)

5. The option can only be exercised at expiration i.e. it is a European type option.

6. No transaction costs i.e. fees on shorting selling etc. 

7. Fractional trading is possible i.e. we can buy/sell 0.x of any given stock. 

Let's discover **B-S** model properties using modeling.

```python
import numpy as np
from scipy.stats import norm

N = norm.cdf

def get_call(S, xs, T, r, sigma):
    d1 = (np.log(S / x_s) + (r + sigma**2 / 2) * T ) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * N(d1) - xs * np.exp(-r * T)* N(d2)

x_s = 100
r = 0.1
T = 1
sigma = 0.3

S = np.arange(60,140,0.1)

calls = [get_call(s, x_s, T, r, sigma) for s in S]

fig, ax = plt.subplots()
fig.set_figwidth(22)
fig.set_figheight(15)

plt.plot(S, calls, 'k-o', label='Call Value')
plt.xlabel('$S_0$, current asset price')
plt.ylabel('Value')
plt.legend()
plt.savefig('curr_price.png')
```
